```python
import dask

def inc(x):
    return x + 1

def double(x):
    return x * 2

def add(x, y):
    return x + y

data = [1,2,3,4,5]

output = []
for x in data:
    a = dask.delayed(inc)(x)
    b = dask.delayed(double)(x)
    c = dask.delayed(add)(a, b)
    output.append(c)

total = dask.delayed(sum)(output)
total.compute()
```

```python
@dask.delayed
def anninc(x):
    return x + 1

@dask.delayed
def anndouble(x):
    return x * 2

@dask.delayed
def annadd(x, y):
    return x + y

data = [1,2,3,4,5]

output = []
for x in data:
    a = anninc(x)
    b = anndouble(x)
    c = annadd(a, b)
    output.append(c)

total = dask.delayed(sum)(output)
total.compute()
```

```python
import matplotlib.pyplot as plt
import numpy as np
steps = ["Dummy", "KNN Training", "KNN Prediction", "KNN Retraining", "KNN Retrained Prediction"]
accuracy = [0.5075, 0.895, 0.963, 0.721, 0.965]
precision = [0.5048, 0.8867, 0.9407, 0.4163, 0.913]
recall = [0.5327, 0.9045, 0.9061, 1.0, 0.8873]

x = np.arange(len(steps))
width = 0.25

fig, ax = plt.subplots(figsize=(12, 6))

plt.plot(x, accuracy, 'o-', color='blue', linewidth=2, label='Accuracy', markersize=8)
plt.plot(x, precision, 's-', color='green', linewidth=2, label='Precision', markersize=8)
plt.plot(x, recall, '^-', color='red', linewidth=2, label='Recall', markersize=8)

for i in range(1, len(steps)):
    plt.axvline(x=i-0.5, color='lightgray', linestyle='--', alpha=0.7)

plt.title('Model Performance Across Different Stages', fontsize=16)
plt.ylabel('Score', fontsize=14)
plt.ylim(0.3, 1.1)
plt.xticks(x, steps, rotation=15, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.legend(loc='lower right', fontsize=12)
plt.annotate('Initial model', xy=(1, accuracy[1]), xytext=(1, accuracy[1]+0.1),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
            ha='center', fontsize=10)
plt.annotate('Performance drop', xy=(3, accuracy[3]), xytext=(3, accuracy[3]-0.15),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
            ha='center', fontsize=10)
plt.annotate('Recovery after retraining', xy=(4, accuracy[4]), xytext=(4, accuracy[4]+0.07),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
            ha='center', fontsize=10)
plt.annotate('Perfect Recall', xy=(3, recall[3]), xytext=(3.3, recall[3]),
            arrowprops=dict(facecolor='red', shrink=0.05, width=1.5, headwidth=8),
            ha='left', fontsize=10)

for i in range(len(steps)-1):
    slope_a = accuracy[i+1] - accuracy[i]
    mid_x = (x[i] + x[i+1]) / 2
    mid_y = (accuracy[i] + accuracy[i+1]) / 2
    plt.annotate(f"{slope_a:.2f}", xy=(mid_x, mid_y), xytext=(mid_x, mid_y + 0.03),
                color='blue', ha='center', fontsize=9)

plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()
```
